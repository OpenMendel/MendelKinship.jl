```julia
using MendelKinship, CSV, DataFrames

#output for openmendel
open_mendel_output = Kinship("control10c.txt")
# open_mendel_output = Kinship("control23a.txt")


# output of Mendel with ran with settings: 
# MODEL = 1
# MIN_MAF = 0.01
# KINSHIP_SOURCE = SNPs_with_everyone
# SNP_SAMPLING_INCREMENT = 1
mendel_output = readdlm("Summary10c.out", header = true)
# mendel_output = readdlm("Summary23a.out", header = true)


# convert everything to dataframe, easier to work with
mendel_output_df = convert(DataFrame, mendel_output[1])
rename!(mendel_output_df, f => t for (f, t) = 
    zip([:x1, :x2, :x3, :x4, :x5, :x6], 
        [:Pedigree1, :Person1, :Pedigree2, 
        :Person2, :expected_kinship, :empiric_kinship]))
```

```julia
using SnpArrays

#make the GRM matrix, to compare with OpenMendel's empiric output later
s = SnpArray("SNP_data10c"; people=212, snps=253141)
GRM = grm(s, method=:GRM, maf_threshold=0.01)
```

### Use Hua's code to create vectorized GRM matrix Φ_snparray matched up with the rest of the data

```julia
ped = readcsv("Ped10c.in")
famid, perid = Int.(ped[2:end, 1]), Int.(ped[2:end, 2])
mendel_famid_1 = Int.(mendel_output_df[:Pedigree1])
mendel_perid_1 = Int.(mendel_output_df[:Person1])
mendel_famid_2 = Int.(mendel_output_df[:Pedigree2])
mendel_perid_2 = Int.(mendel_output_df[:Person2])
Φ_theoretical = mendel_output_df[:expected_kinship]
Φ_empirical = mendel_output_df[:empiric_kinship]

function index_in_ped(fid, pid, fid_list, pid_list)
    for i in 1:length(fid_list)
        (fid == fid_list[i] && pid == pid_list[i]) && return i
    end
    return 0
end

Φ_snparray = similar(Φ_empirical)
for iter = 1:length(Φ_empirical)
    i = index_in_ped(mendel_famid_1[iter], mendel_perid_1[iter], famid, perid)
    j = index_in_ped(mendel_famid_2[iter], mendel_perid_2[iter], famid, perid)
    Φ_snparray[iter] = GRM[i, j]
end
```

### Compare mendel, open mendel, and GRM matrix from SnpArrays by constructing convenient dataframe

```julia
compare = DataFrame(
    Person1 = String[], 
    Person2 = String[], 
    Pedigree1 = String[],
    Pedigree2 = String[],
    Mendel_empiric= Float64[], 
    OpenMendel_empiric=Float64[],
    Φ_SnpArray=Float64[],
    Mendel_theoretical=Float64[],
    OpenMendel_theoretical=Float64[])

m = size(mendel_output_df, 1)
for i in 1:m
    cur_row = mendel_output_df[i, :]
    if cur_row[6][1] == "--"
        continue
    end
    
    person1 = string(Int64(cur_row[2][1]))
    person2 = string(Int64(cur_row[4][1]))
    person1_ped = string(Int64(cur_row[1][1]))
    person2_ped = string(Int64(cur_row[3][1]))


    y = copy(open_mendel_output)
    
    #find rows with Person1 = person1
    y = y[y[:, 3] .== person1, :]

    #find rows with Person2 = person2
    y = y[y[:, 4] .== person2, :]

    push!(compare, [person1, person2, person1_ped, person2_ped, 
        cur_row[6][1], y[6][1], Φ_snparray[i], cur_row[5][1], y[5][1]])
end
```

```julia
# find large deviations
huge_deviations = DataFrame(
    Person1 = String[], 
    Person2 = String[], 
    Pedigree1 = String[],
    Pedigree2 = String[],
    Mendel_empiric= Float64[], 
    OpenMendel_empiric=Float64[],
    Φ_SnpArray=Float64[],
    mendel_theoretical=Float64[],
    openmendel_theoretical=Float64[])

m = size(compare, 1)

for i in 1:m
    cur_row = compare[i, :]
    cur_row = [cur_row[1][1], cur_row[2][1], 
               cur_row[3][1], cur_row[4][1],
               cur_row[5][1], cur_row[6][1],
               cur_row[7][1], cur_row[8][1], 
               cur_row[9][1]]
    
    if abs(cur_row[8] - cur_row[6]) >= 0.05
        push!(huge_deviations, cur_row)
    end
end
```
